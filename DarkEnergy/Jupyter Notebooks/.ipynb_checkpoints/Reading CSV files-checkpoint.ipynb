# Reading a flat file

```python
import csv
with open('/home/mmarionn/Documents/ETHZ/StatExpPhy_1617/h1new/h1_new_data.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    n=0
    for row in reader:
        print(row['nrun'], row['reh'])
        n+=1
        if n>10: break
```
